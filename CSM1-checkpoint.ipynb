```python
def f1_metric(model, X_train, y_train):
    '''
    This function reports f1-score metric for the class specified by 'Positive Label' (or '1')
    '''
    return f1_score(y_train, model.predict(X_train), average='binary')
```

```python
def drop_column_importance(model, X_train, y_train, random_state=0):
    # list containing feature importances
    importances = []
    # Clone the model
    model_clone = clone(model)
    # Set random_state for comparability
    model_clone.random_state = random_state
    # Train the model
    model_clone.fit(X_train, y_train)
    # Create the cross-validation object using StratifiedKFold 
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
    # Score the benchmark model using cross-validation
    benchmark_score = cross_val_score(model_clone, X_train, y_train, cv=cv, scoring=f1_metric).mean()

    
    # Iterate over all features and store feature importance
    for col in X_train.columns:
        # Clone the model
        model_clone = clone(model)
        # Set random_state for comparability
        model_clone.random_state = random_state
        # Traint the model on the the dataset with a single feature removed
        model_clone.fit(X_train.drop(col, axis = 1), y_train)
        # Score the droped-column model
        drop_column_score = cross_val_score(model_clone, X_train.drop(col,axis=1), y_train, cv=cv, scoring=f1_metric).mean()
        # Store feature importance which is defined as the difference between the benchmark and the new model score
        importances.append(benchmark_score - drop_column_score)
        
    # Return the features along with their importances in the form of a dataframe
    importances_df_train = pd.DataFrame({'feature': X_train.columns, 'feature importance': importances}) \
                     .sort_values('feature importance', ascending = False).reset_index(drop = True)
           
    return importances_df_train
```

```python
def drop_column_importance_plot(model, X_train, y_train):
    # Call drop-column feature importance function
    df_train_drop_column = drop_column_importance(model, X_train, y_train, random_state=0)
    # Rename columns
    df_train_drop_column.columns = ['Feature', 'Feature Importance']
    
    # Plot barchart
    plt.figure(figsize=(12,10))
    sns.barplot(data=df_train_drop_column, x='Feature Importance', y='Feature', orient='h', color='royalblue')
    plt.title('Drop Column Feature Importance', fontsize=20)
    plt.show()
```

```python
def discretization_report(df_train, clf):
    '''
    This function finds the optimal combination of n_bins and strategy for continuous features discretization
    '''
    # Define continuous features to perform discretization on
    cols_to_discretize = ['Monthly_Balance', 'Total_EMI_per_month', 'Age', 'Annual_Income']

    # Define the features (X) and the output labels (y) 
    X = df_train[cols_to_discretize]
    y = df_train['Credit_Score']

    # Split dataset into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)

    # Define the grid search parameters
    param_grid = {'discretizer__strategy': ['uniform', 'quantile', 'kmeans'],
                  'discretizer__n_bins': np.arange(2,11)}

    # Define the KBinsDiscretizer and OneHotEncoder and ComplementNB objects
    discretizer = KBinsDiscretizer(encode='ordinal')
    onehot = OneHotEncoder(handle_unknown='ignore', drop='first')

    # Create the pipeline
    pipeline = Pipeline([('discretizer', discretizer), ('onehot', onehot), ('clf', clf)])

    # Create the cross-validation object using StratifiedKFold to ensure the class distribution is the same across all the folds
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

    # Create the GridSearchCV object
    grid_search = GridSearchCV(pipeline, param_grid, cv=cv, scoring='f1')
    
    # Fit the GridSearchCV object to the training data
    grid_search.fit(X_train, y_train)

    # Print the best parameters and the best score
    print("Best discretization parameters:", grid_search.best_params_)
    print("Best score:", grid_search.best_score_)
    
    # Return optimal values for n_bins and strategy
    return grid_search.best_params_['discretizer__n_bins'], grid_search.best_params_['discretizer__strategy']
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score
from copy import deepcopy

def drop_column_importance_plot(model, X_train, y_train, scoring_func=f1_score, average='macro'):
    base_model = deepcopy(model)
    base_model.fit(X_train, y_train)
    baseline_score = scoring_func(y_train, base_model.predict(X_train), average=average)

    importances = []

    for col in X_train.columns:
        X_dropped = X_train.drop(columns=[col])
        model_copy = deepcopy(model)
        model_copy.fit(X_dropped, y_train)
        drop_score = scoring_func(y_train, model_copy.predict(X_dropped), average=average)
        importance = baseline_score - drop_score
        importances.append((col, importance))

    importances.sort(key=lambda x: x[1], reverse=True)

    # Plot
    cols, scores = zip(*importances)
    plt.figure(figsize=(10, 6))
    plt.barh(cols, scores)
    plt.xlabel("Importance (F1 score drop)")
    plt.title("Drop Column Importance")
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.metrics import roc_auc_score, roc_curve, accuracy_score, classification_report

def model_evaluation(model, X_train, X_test, y_train, y_test, pos_label=None, model_name="Model"):
    # Predict probabilities
    y_prob_train = model.predict_proba(X_train)
    y_prob_test  = model.predict_proba(X_test)

    # If binary classification
    if y_prob_train.shape[1] == 2:
        y_prob_train = y_prob_train[:, 1]
        y_prob_test = y_prob_test[:, 1]
        auc_train = roc_auc_score(y_train, y_prob_train)
        auc_test  = roc_auc_score(y_test,  y_prob_test)
    else:
        # Multiclass AUC
        auc_train = roc_auc_score(y_train, y_prob_train, multi_class='ovr')
        auc_test  = roc_auc_score(y_test,  y_prob_test,  multi_class='ovr')

    print(f"{model_name} AUC (train): {auc_train:.4f}")
    print(f"{model_name} AUC (test):  {auc_test:.4f}")

    # Accuracy
    y_pred_test = model.predict(X_test)
    acc_test = accuracy_score(y_test, y_pred_test)
    print(f"{model_name} Accuracy (test): {acc_test:.4f}")

    # Classification report
    print(f"\nClassification Report:\n{classification_report(y_test, y_pred_test)}")

```

```python
import pandas as pd
from sklearn.metrics import accuracy_score

def drop_column_importance(model, X_train, y_train, baseline_score=None):
    """
    Measures feature importance by drop-column method.
    """
    if baseline_score is None:
        model.fit(X_train, y_train)
        baseline_score = accuracy_score(y_train, model.predict(X_train))

    importances = []
    for col in X_train.columns:
        X_train_dropped = X_train.drop(columns=[col])
        model.fit(X_train_dropped, y_train)
        drop_score = accuracy_score(y_train, model.predict(X_train_dropped))
        importance = baseline_score - drop_score
        importances.append({'feature': col, 'feature importance': importance})

    return pd.DataFrame(importances).sort_values(by='feature importance', ascending=False)
```
